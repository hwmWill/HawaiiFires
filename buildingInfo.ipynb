```python
import pandas as pd
```

```python
df = pd.read_csv('./data/comdat23/comdat23.txt', delimiter='\t',
                 header=None)
```

```python
df.head()
```

```python
df['island']      = df[0].str[0]
df['zone']        = df[0].str[1]
df['section']     = df[0].str[2]
df['plat']        = df[0].str[3:6]
df['parcel']      = df[0].str[6:9]
df['cpr']         = df[0].str[9:13]
df['tax_yr']      = df[0].str[13:18]
df['card']        = df[0].str[18:23]
df['ident_units'] = df[0].str[23:31]
df['units']       = df[0].str[31:36]
df['strc_type']   = df[0].str[36:39].str.strip()
df['grade']       = df[0].str[39:42].str.strip()
df['strc_desc']   = df[0].str[42:72].str.strip()
df['yr_built']    = df[0].str[72:77]
df['effec_yr']    = df[0].str[77:82]
df['impr_name']   = df[0].str[82:112].str.strip()
df['bldng_val']   = df[0].str[112:122]
# df['bldng_area']  = df[0].str[122:]
```

```python
df.head()
```

```python
df.replace(r'^\s*$', None, regex=True, inplace = True)
```

```python
df['tax_yr']      = df.tax_yr.astype('float')
df['card']        = df.card.astype('float')
df['ident_units'] = df.ident_units.astype('float')
df['units']       = df.units.astype('float')
df['yr_built']    = df.yr_built.astype('float')
df['effec_yr']    = df.effec_yr.astype('float')
df['bldng_val']   = df.bldng_val.astype('float')
```

```python
df.drop(columns=0, inplace=True)
```

```python
df.head().T
```

```python
df['tmk_text'] = df.island + df.zone + df.section + df.plat + df.parcel
```
